### 6.2
#### b)

```python
import numpy as np
import matplotlib.pyplot as plt

# Setup variables
t = -0.2
eta = np.linspace(-2, 2)
eta_l, eta_g = -2*np.sqrt(-t), 2*np.sqrt(-t) # limits to convex region
cvx = np.linspace(eta_l, eta_g, sum(np.logical_and(eta >= eta_l, eta <= eta_g))) # convex region

# Define curve functions
f = lambda x: (3/8)*x**4 + 3*t*x**2 - (1+4*t)*x # original (uncorrected) f 
fbar = lambda x: -(1+4*t)*x - 6*t**2 # bitangent line to f
fp = lambda x: (3/2)*x**3 + 6*t*x - (1+4*t) # derivative of f

# Data
fo = f(eta) # original f as data
fc = np.concatenate((fo[eta < eta_l], fbar(cvx), fo[eta > eta_g])) # corrected f as data

# Plot curves
plt.plot(eta, fo, 'b-', label="$f$")
plt.plot(eta, fc, 'b--', label=r"$\bar{f}$")
plt.title(r"$f$-curves vs. $\eta$")
plt.xlabel(r"$\eta$")
plt.ylabel("$f$")
plt.legend()
plt.grid(True)
```

```python
# Calculate pressure
p = -fp(eta) # using derivative of f
pc = np.concatenate((p[eta < eta_l], \
                     (1+4*t)*np.ones(len(cvx)), \
                    p[eta > eta_g])) # convex region

# Plot pressure
plt.plot(eta, p, 'b-', label="Uncorrected")
plt.plot(eta, pc, 'b--', label="Corrected")
plt.title(r"$p$ vs. $\eta$")
plt.xlabel(r"$\eta$")
plt.ylabel("$p$")
plt.legend()
plt.grid(True)
```

```python
# Perform Legendre Transform
g = fo + eta*p # transform values
gc = fc + eta*pc # transform values in convex region
i = np.argsort(-p) # find sort indices of derivative

# Plot g vs. p (ordered)
plt.figure()
plt.plot(-p[i], g[i])
plt.title(r"$G$ vs. $p$")
plt.xlabel(r"$p$")
plt.ylabel(r"$G$")
plt.grid(True)

plt.figure()
plt.plot(-pc[i], gc[i])
plt.title(r"$G$ vs. $p$ (corrected)")
plt.xlabel(r"$p$")
plt.ylabel(r"$G$")
plt.grid(True)
```

```python
a = 1
M = np.linspace(-1.5, 1.5)
h = M/(2*a)
Fo = 0

F = -Fo - a*h**2
A = -Fo + M**2/(4*a)

# Make plots
plt.figure()
plt.plot(h, F)
plt.title("$F$ vs. $h$")
plt.xlabel("$h$")
plt.ylabel("$F$")
plt.grid(True)

plt.figure()
plt.plot(M, A)
plt.title("$A$ vs. $M$")
plt.xlabel("$M$")
plt.ylabel("$A$")
plt.grid(True)
```

#### b)

```python
F = -a*abs(h) - h**2/2
Mnew = np.concatenate((a-h[h<0], -a-h[h>0]))

# Make plots
plt.figure()
plt.plot(h, F, label="$F(h)$")
plt.plot(h[h<0], Mnew[h<0], 'r-', label="$M(h)$")
plt.plot(h[h>0], Mnew[h>0], 'r-')
plt.xlabel("$h$")
plt.legend()
plt.grid(True)

# Plot h(M)
plt.figure()
plt.plot(Mnew[h<0], h[h<0], 'r-')
plt.plot(Mnew[h>0], h[h>0], 'r-')
plt.title("$h$ vs. $M$")
plt.xlabel("$M$")
plt.grid(True)
```

```python
# Legendre Transform
i = np.argsort(Mnew) # sort based on M
A = F + h*Mnew # transform values

plt.figure()
plt.plot(Mnew[h<0], A[h<0], 'r-')
plt.plot(Mnew[h>0], A[h>0], 'r-')
plt.title("$A$ vs. $M$")
plt.xlabel("$M$")
plt.ylabel("$A$")
plt.grid(True)
```
